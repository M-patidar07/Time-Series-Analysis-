```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the dataset
df = pd.read_csv('car-sales-india-2021-1.csv')

```

```python
# Extract sales data for each year
years = [2019, 2020, 2021]
```

```python
# Filter Indian and Non-Indian companies
df_indian = df[df['Country'] == 'India']
df_non_indian = df[df['Country'] != 'India']

```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
# Plot for each year
plt.figure(figsize=(18, 6))
for i, year in enumerate(years, 1):
    plt.subplot(1, 3, i)
    plt.plot(df_indian[df_indian['Year'] == year].iloc[:, 3:15].sum(axis=0), label='Indian', marker='o')
    plt.plot(df_non_indian[df_non_indian['Year'] == year].iloc[:, 3:15].sum(axis=0), label='Non-Indian', marker='o')
    plt.xlabel('Month')
    plt.ylabel('Sales')
    plt.title(f'Sales Comparison for {year}')
    plt.xticks(range(12), df.columns[3:15], rotation=45)
    plt.legend()
    plt.grid(True)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
