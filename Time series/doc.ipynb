```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generating example data
np.random.seed(0)
dates = pd.date_range(start='2019-01-01', end='2021-12-01', freq='M')
sales = 1000 + np.cumsum(np.random.randn(len(dates)))
```

```python
# Creating a pandas DataFrame
df = pd.DataFrame({'Date': dates, 'Sales': sales})
df.set_index('Date', inplace=True)
```

```python
# Plotting the time series data
plt.figure(figsize=(8, 6))
plt.plot(df.index, df['Sales'], color='blue', label='Sales')
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generating example data
np.random.seed(0)
dates = pd.date_range(start='2019-01-01', end='2021-12-01', freq='M')
temperature = 20 + 10 * np.sin(np.arange(len(dates)) * (2 * np.pi / 12)) + np.random.randn(len(dates))

# Creating a pandas DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperature})
df.set_index('Date', inplace=True)

```

```python
# Plotting the time series data
plt.figure(figsize=(8, 6))
plt.plot(df.index, df['Temperature'], color='red', label='Temperature')
plt.title('Monthly Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generating example data
np.random.seed(0)
dates = pd.date_range(start='2010-01-01', end='2023-12-01', freq='M')
stock_prices = 1000 + 200 * np.sin(np.arange(len(dates)) * (2 * np.pi / 12)) + np.random.randn(len(dates))

# Creating a pandas DataFrame
df = pd.DataFrame({'Date': dates, 'Stock_Price': stock_prices})
df.set_index('Date', inplace=True)

```

```python
# Plotting the time series data
plt.figure(figsize=(8, 6))
plt.plot(df.index, df['Stock_Price'], color='green', label='Stock Price')
plt.title('Monthly Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```
